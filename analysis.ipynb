```python
# Import the yfinance. If you get module not found error the run !pip install yfianance from your Jupyter notebook
import yfinance as yf

# Get the data for the stock AAPL
data = yf.download('0002.hk',start="2000-01-01",end="2020-09-30")

# Import the plotting library
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the close price of the AAPL
data['Adj Close'].plot()
plt.show()
```

```python
# Define the ticker list
import pandas as pd
#tickers_list = ['0002.hk', '0003.hk', '2800.hk', '0066.hk','1299.hk','0052.hk','0345.hk']
tickers_list = ['0003.hk', '0066.hk','0868.hk','0002.hk','0868.HK','1299.hk','0823.hk','0052.hk','0388.hk','1038.HK']
# Import pandas
data = pd.DataFrame(columns=tickers_list)

# Fetch the data
import yfinance as yf
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, start="2010-01-01",end="2020-04-30")['Adj Close']

# Print first 5 rows of the data
data.head()
```

```python
# Plot all the close prices
((data.pct_change()+1).cumprod()).plot(figsize=(16, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Adjusted Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
